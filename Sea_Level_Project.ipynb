```python
import numpy as np 
import pandas as pd 
import datetime as dt
import sklearn as sk
import itertools
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates
import matplotlib as mpl
mpl.style.use('seaborn')
from matplotlib.offsetbox import AnchoredText
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
plt.style.use('ggplot')
%matplotlib inline 
```

```python
df = pd.read_csv('csiro_alt_gmsl_yr_2015_csv.csv')
x = df["Time"]
y = df['GMSL']
fig, ax = plt.subplots(figsize=(12, 6))
plt.plot(x, y)
plt.show()
```

```python
def draw_plot():
    df = pd.read_csv('Sealevel.csv')

    plt.scatter(df['Time'], df['GMSL'])

    lineA = linregress(df['Time'], df['GMSL'])
    xA = np.arange(df['Time'].min(),2100,1)
    yA = xA*lineA.slope + lineA.intercept

    plt.plot(xA,yA)

    df_2000 = df[df['Time'] >= 2000]

    lineB = linregress(df_2000['Time'], df_2000['GMSL'])
    xB = np.arange(2000,2050,1)
    yB = xB*lineB.slope + lineB.intercept

    plt.plot(xB,yB)

    # Add labels and title
    plt.xlabel('Time')
    plt.ylabel('Sea Level (inches)')
    plt.title('Rise in Sea Level')
```

```python
draw_plot()
```
